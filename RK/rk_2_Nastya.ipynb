```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
data = pd.read_csv('states_all_extended.csv')
data.head()
```

```python
data_res = data.dropna(thresh=int(len(data) * .8), axis=1)
```

```python
data_res.info()
```

```python
data_res = data_res.drop(columns=['PRIMARY_KEY'])
data_res = data_res.dropna(axis=0)
data_res['STATE'] = data_res['STATE'].astype('category')
data_res.isnull().sum()
```

```python
data_res['STATE'].unique()
```

```python
data_res.info()
```

```python
category_columns = data_res.select_dtypes(include=['category']).columns
category_columns
```

```python
from sklearn.preprocessing import OneHotEncoder


ohe = OneHotEncoder()
encoded_columns = ohe.fit_transform(data_res[category_columns])
```

```python
import numpy as np


data_res[np.concatenate(ohe.categories_)] = encoded_columns.toarray()
data_res.drop(['STATE'], axis=1, inplace=True)
data_res = data_res[[x for x in data_res.columns if x != 'PK_A_A']+['PK_A_A']]
data_res
```

```python
TARGET_KEY = 'PK_A_A'
X_data, y_data = data_res.drop(columns=[TARGET_KEY]), data_res[[TARGET_KEY]]
X_test, X_train, y_test, y_train = train_test_split(X_data, y_data)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_pred
```

```python
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
gb_pred
```

Для сравнения качества моделей используем метрики средней абсолютной ошибки и коэффициента детерминации. Первый покажет, насколько предсказанные значения отклоняются от настоящих. Второй покажет, насколько точно модель предсказывает значения в целом.

```python
lr_mae = mean_absolute_error(y_test, lr_pred)
gb_mae = mean_absolute_error(y_test, gb_pred)
```

```python
lr_r2 = r2_score(y_test, lr_pred)
gb_r2 = r2_score(y_test, gb_pred)
```

```python
pd.DataFrame.from_dict({
    'LR': {
        'MAE': round(lr_mae, 5),
        'R2': f'{round(lr_r2 * 100, 3)}%'
    },
    'GB': {
        'MAE': round(gb_mae, 5),
        'R2': f'{round(gb_r2 * 100, 3)}%'
    }
})
```

По результатам можно сказать, что модели можно признать приемлемыми. Также можно заметить, что линейная регрессия справляется с задачей немного лучше, выдавая достаточно хоршие результаты.
